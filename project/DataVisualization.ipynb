```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load the CSV files into dataframes
df1 = pd.read_csv('/content/drive/MyDrive/MADE/traffic_collision_data.csv')  # Your traffic collision data
df2 = pd.read_csv('/content/drive/MyDrive/MADE/weather_data.csv')  # Your weather data

# Step 2: Ensure the date columns are in datetime format
df1['date_occ'] = pd.to_datetime(df1['date_occ'])
df2['datetime'] = pd.to_datetime(df2['datetime'])

# Step 3: Merge the dataframes on the common date column
merged_df = pd.merge(df1, df2, left_on='date_occ', right_on='datetime')

# Step 4: Group by date_occ and calculate the total count, and average temp and humidity
count_per_day = merged_df.groupby('date_occ').agg(
    total_count=('date_occ', 'size'),  # Count of entries per date
    avg_temp=('temp', 'mean'),  # Average temperature for the day
    avg_humidity=('humidity', 'mean')  # Average humidity for the day
).reset_index()

# Step 5: Calculate Correlations
correlation_temp_count = count_per_day['avg_temp'].corr(count_per_day['total_count'])
correlation_humidity_count = count_per_day['avg_humidity'].corr(count_per_day['total_count'])

print("Correlation between temperature and total count:", correlation_temp_count)
print("Correlation between humidity and total count:", correlation_humidity_count)

# Step 6: Plotting the correlations
plt.figure(figsize=(12, 6))

# Plot Temp vs Total Count
plt.subplot(1, 2, 1)
sns.scatterplot(data=count_per_day, x='avg_temp', y='total_count')
plt.title('Average Temperature vs Total Count Per Day')
plt.xlabel('Average Temperature (C)')
plt.ylabel('Total Count')

# Plot Humidity vs Total Count
plt.subplot(1, 2, 2)
sns.scatterplot(data=count_per_day, x='avg_humidity', y='total_count')
plt.title('Average Humidity vs Total Count Per Day')
plt.xlabel('Average Humidity (%)')
plt.ylabel('Total Count')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
collision_df = pd.read_csv('/content/drive/MyDrive/MADE/traffic_collision_data.csv')
weather_df = pd.read_csv('/content/drive/MyDrive/MADE/weather_data.csv')

# Ensure date columns are in datetime format
collision_df['date_occ'] = pd.to_datetime(collision_df['date_occ'])
weather_df['datetime'] = pd.to_datetime(weather_df['datetime'])

# Merge the datasets on the date column
merged_df = pd.merge(collision_df, weather_df, left_on='date_occ', right_on='datetime')

# Verify merged data
if merged_df.empty:
    raise ValueError("Merged DataFrame is empty. Check date formats or overlapping dates.")

# Calculate the correlation matrix
correlation_matrix = merged_df[['collision_count', 'temp', 'humidity', 'windspeed', 'visibility']].corr()

# Plot the correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={"shrink": .8})
plt.title('Correlation between Collision Counts and Weather Parameters', fontsize=16)
plt.show()
```
